```python
import pandas as pd 
```

# Import Data 

## Import Player Data 

### Past Player Data

```python
df_players = pd.read_csv("../Data/player_data.csv")
df_players.drop("Unnamed: 0", axis = 1, inplace = True) 
df_players
```

```python
df_label = df_players[["season", "name", "is_mvp"]]
df_players.drop("is_mvp", axis=1, inplace=True)
df_label
```

### 2023 Player Data

```python
df_players_2023 = pd.read_csv("../Data/player_data_2023.csv")
df_players_2023.drop("Unnamed: 0", axis = 1, inplace = True) 
df_players_2023
```

## Import Team Data

### Past Team Data

```python
df_teams = pd.read_csv("../Data/team_data.csv")
df_teams.drop("Unnamed: 0", axis=1, inplace=True) 
df_teams
```

### 2023 Team Data 

```python
df_teams_2023 = pd.read_csv("../Data/team_data_2023.csv")
df_teams_2023.drop("Unnamed: 0", axis=1, inplace=True) 
df_teams_2023
```

## Import Media Narrative Data

`MVP_count.csv` is designed to capture the number of time a player have won MVP and whether he has won it the previous year or previous 2 years. It is generally agreed upon that players who win their first MVP the previous year tend to win another MVP, whilst players who have won MVP 2 years in a row are highly unlikely to win a third one (sometimes termed "voter fatigue"). This dataset is designed to capture the aspect of "voter fatigue". 

```python
df_mvp_count = pd.read_csv("../Data/MVP_count.csv")
df_mvp_count.rename(columns={"Name":"name", "Year":"season", "value":"num_mvp"}, inplace=True) 
df_mvp_count
```

# Merge Data

## Merge Past Data

```python
df_final = pd.merge(df_players, df_teams, on=["season", "team"], how="left") 
df_final.drop(["conference", "W", "L"], axis=1, inplace=True)
df_final
```

```python
df_final = pd.merge(df_final, df_mvp_count, on=["season", "name"], how="left") 
df_final = df_final.fillna(0)
df_final[["num_mvp", "won_last_yr", "won_2_yrs_before"]] = df_final[["num_mvp", "won_last_yr", "won_2_yrs_before"]].astype(int)
df_final
```

## Merge 2023 Data

```python
df_final_2023 = pd.merge(df_players_2023, df_teams_2023, on=["season", "team"], how="left") 
df_final_2023.drop(["conference", "W", "L"], axis=1, inplace=True)
df_final_2023
```

# Add label 

```python
df_final = pd.merge(df_final, df_label, on=["season", "name"], how="left")
df_final
```

```python
df_final_2023 = pd.merge(df_final_2023, df_mvp_count, on=["season", "name"], how="left") 
df_final_2023 = df_final_2023.fillna(0)
df_final_2023[["num_mvp", "won_last_yr", "won_2_yrs_before"]] = df_final_2023[["num_mvp", "won_last_yr", "won_2_yrs_before"]].astype(int)
df_final_2023
```
