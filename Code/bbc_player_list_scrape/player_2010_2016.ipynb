```python
df_2016 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2016_player_data.csv')
df_2015 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2015_player_data.csv')
df_2014 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2014_player_data.csv')
df_2013 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2013_player_data.csv')
df_2012 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2012_player_data.csv')
df_2011 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2011_player_data.csv')
df_2010 = pd.read_csv('/Users/annabelitong/desktop/annabel/uni/ds105l/group_project/DS105L-Project/Data/2010-2016/2010_player_data.csv')
```

```python
def get_player_news(one_season):
    base_url = 'https://www.bbc.co.uk/search'
    date_l = []
    url_l = []
    title_l = []
    name_l = []
        
    for player in one_season:
            print(player)
            results = []
            page_number = 1
            title_1 = []

            url_player = player.replace(" ", "-")
            while True:
                if page_number > 29:
                     break
                params = {'q': url_player  + '+mvp', 'd': 'SPORT_PS', 'page': str(page_number)}

                try:
                    response = requests.get(base_url, params=params)
                    soup = BeautifulSoup(response.text, 'html.parser')
                    event_cards = soup.find_all('div', attrs={"data-testid": "default-promo", "class":"ssrcss-53phst-Promo ett16tt0"})

                    if not event_cards:
                        break

                    for card in event_cards:
                        title = card.find('p', attrs={'class': "ssrcss-6arcww-PromoHeadline e1f5wbog5"}).text.strip()
                        title_1.append(title)
                        date = card.find('span', attrs={'class': 'ssrcss-1if1g9v-MetadataText ecn1o5v1'}).text.strip()
                        date_l.append(date)
                        url = card.find('a').get('href')
                        url_l.append(url)
                        results.append({'title': title, 'date': date, 'url': url})

                    page_number += 1
                    print(page_number)

                except Exception as e:
                    break
            name_l += [player] * len(title_1)
            title_l += title_1
    one_season_df = pd.DataFrame({"title": title_l, "date":date_l, 'url': url_l,"player": name_l })
    return one_season_df
        

    
```

```python
bbc_2016 = get_player_news(list(df_2016["name"]))
bbc_2015 = get_player_news(list(df_2015["name"]))
bbc_2014 = get_player_news(list(df_2014["name"]))
bbc_2013 = get_player_news(list(df_2013["name"]))
bbc_2012 = get_player_news(list(df_2012["name"]))
bbc_2011 = get_player_news(list(df_2011["name"]))
bbc_2010 = get_player_news(list(df_2010["name"]))
```
