```python
import pandas as pd 
```

# Importing Data

## Importing Box Score Data 

```python
df_all = pd.read_csv("../Data/merged_player_data_2010_2023.csv") 
df_all.drop(df_all.columns[df_all.columns.str.contains("Unnamed")], axis = 1, inplace = True) 
df_all["year"] = df_all["year"].apply(lambda x: int(x[:4])) 
df_all.drop("positions", axis = 1, inplace = True)
df_all
```

```python
targets = ["year", "slug", "name", "team", "age", "games_played"] 
targets = targets[::-1]
cols = list(df_all.columns)
for col in targets: 
    cols.insert(0, cols.pop(cols.index(col))) 
df_all = df_all.loc[:, cols]
df_all.rename(columns = {"year": "season"}, inplace = True)
df_all
```

## Importing Advanced Statistics 

```python
df_advanced = pd.read_csv("../Data/merged_advanced_player_data_2010_2023.csv") 
df_advanced.drop(df_advanced.columns[df_advanced.columns.str.contains("Unnamed")], axis = 1, inplace = True) 
df_advanced["year"] = df_advanced["year"].apply(lambda x: int(x[:4])) 
df_advanced.drop(["positions", "minutes_played", "is_combined_totals"], axis = 1, inplace = True)
df_advanced
```

```python
targets = ["year", "slug", "name", "team", "age", "games_played"] 
targets = targets[::-1]
cols = list(df_advanced.columns)
for col in targets: 
    cols.insert(0, cols.pop(cols.index(col))) 
df_advanced = df_advanced.loc[:, cols]
df_advanced.rename(columns = {"year": "season"}, inplace = True)
df_advanced
```

# Clean up duplicates due to trades 

```python
id_cols = df_all.iloc[:, :5].columns 
stats_cols = df_all.iloc[:, 5:].columns 
df_group = df_all.groupby(["season", "slug", "name"]) 
df_str = df_group.agg({"team": lambda x: ", ".join(x), "age": "min"})
df_int = df_group[stats_cols].agg("sum")
df_no_dup = df_str.merge(df_int, left_index = True, right_index = True).reset_index()
df_no_dup
```

```python
id_cols = df_advanced.iloc[:, :5].columns 
stats_cols = df_advanced.iloc[:, 5:].columns 
df_group = df_advanced.groupby(["season", "slug", "name"]) 
df_str = df_group.agg({"team": lambda x: ", ".join(x), "age": "min"})
df_int = df_group[stats_cols].agg("sum")
df_no_dup_advanced = df_str.merge(df_int, left_index = True, right_index = True).reset_index()
df_no_dup_advanced
```

# Per game Data

```python
df_id = df_no_dup.iloc[:, :6]
df_per_game = df_no_dup.iloc[:, 6:].div(df_no_dup["games_played"], axis = 0) 
df_per_game = df_id.merge(df_per_game, left_index = True, right_index = True)
df_per_game
```

## Create %games_played 

```python
df_num_games_2020 = pd.read_csv("../Data/num_games_2020.csv") 
df_num_games_2023 = pd.read_csv("../Data/num_games_2023.csv") 
```

```python
df_num_games_2020 = df_num_games_2020.iloc[1:, 1:3] 
df_num_games_2020.rename(columns = {"Unnamed: 1": "team", "Unnamed: 2": "record"}, inplace = True) 
df_num_games_2020["total_games"] = df_num_games_2020["record"].apply(lambda x: int(x.split("-")[0]) + int(x.split("-")[1])) 
df_num_games_2020.drop("record", axis = 1, inplace = True) 
df_num_games_2020["team"] = df_num_games_2020["team"].apply(lambda x: x.upper())
df_num_games_2020
```

```python
df_num_games_2023 = df_num_games_2023.iloc[1:, 1:3] 
df_num_games_2023.rename(columns = {"Unnamed: 1": "team", "Unnamed: 2": "record"}, inplace = True) 
df_num_games_2023["total_games"] = df_num_games_2023["record"].apply(lambda x: int(x.split("-")[0]) + int(x.split("-")[1])) 
df_num_games_2023.drop("record", axis = 1, inplace = True) 
df_num_games_2023["team"] = df_num_games_2023["team"].apply(lambda x: x.upper())
df_num_games_2023
```

```python
def fill_total_games(df): 
    if df["season"] == 2021: 
        return 72 
    elif df["season"] == 2012: 
        return 66
    elif df["season"] == 2020: 
        try: 
            return df_num_games_2020[df_num_games_2020["team"] == df["team"]]["total_games"].values[0] 
        except IndexError: #Handle traded players by returning the average num_games played by all teams
            return int(df_num_games_2020["total_games"].mean())
    elif df["season"] == 2023: 
        try: 
            return df_num_games_2023[df_num_games_2023["team"] == df["team"]]["total_games"].values[0] 
        except IndexError: #Handle traded players by returning the average num_games played by all teams
            return int(df_num_games_2023["total_games"].mean())
    else: 
        return 82
```

```python
df_per_game["total_games"] = df_per_game.apply(lambda x: fill_total_games(x), axis = 1)
df_per_game["games_played"] = df_per_game["games_played"]/df_per_game["total_games"] 
df_per_game = df_per_game.rename(columns = {"games_played": "%games_played"}).drop("total_games", axis = 1)
df_per_game
```

```python
df_advanced["total_games"] = df_advanced.apply(lambda x: fill_total_games(x), axis = 1)
df_advanced["games_played"] = df_advanced["games_played"]/df_advanced["total_games"] 
df_advanced = df_advanced.rename(columns = {"games_played": "%games_played"}).drop("total_games", axis = 1)
df_advanced
```

## Create field_goal%, free_throw%, and 3pt%

```python
df_per_game["field_goal%"] = df_per_game["made_field_goals"]/df_per_game["attempted_field_goals"] 
df_per_game["free_throw%"] = df_per_game["made_free_throws"]/df_per_game["attempted_free_throws"] 
df_per_game["3pt"] = df_per_game["made_three_point_field_goals"]/df_per_game["attempted_three_point_field_goals"] 
df_per_game
```

## Eliminate Counting Stats 

```python
counting_stats = ["win_shares", "offensive_win_shares", "defensive_win_shares"]
df_advanced.drop(columns = counting_stats, axis = 1, inplace = True) 
df_advanced
```

# Merge box score and advanced stats to create `df_players` 

```python
df_players = df_per_game.merge(df_advanced, how = "inner", on = ["season", "slug", "name", "team", "age", "%games_played"])
df_players
```

```python
df_players.columns
```

## Separate 2023 data from the rest (2010 – 2022)

```python
df_players_2023 = df_players[df_players["season"] == 2023] 
df_players = df_players[df_players["season"] != 2023] 
```

## Add label for MVP winners 

```python
df_mvp = pd.read_csv("../Data/mvp_winners_history.csv") 
df_mvp = df_mvp.iloc[1:, [0, 2, -1]] 
df_mvp.rename(columns = {"Unnamed: 0": "season", "Unnamed: 2": "name", "-additional": "slug"}, inplace = True) 
df_mvp["season"] = df_mvp["season"].apply(lambda x: int(x[:-3]) + 1)
df_mvp
```

```python
def label_mvp(df): 
    if (df["slug"] == df_mvp[df_mvp["season"] == df["season"]]["slug"].values[0]): 
        return 1 
    else: 
        return 0 
```

```python
df_players["is_mvp"] = df_players.apply(lambda x: label_mvp(x), axis = 1) 
df_players
```
