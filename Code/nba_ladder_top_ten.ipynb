```python
all_top_ten = []
#for links in list(ladder_df["link"]):
for links in list(ladder_df["link"]):
    try:
        r = requests.get(links) 
        soup = BeautifulSoup(r.content, "html.parser") 
        title = soup.find("h1").get_text()
        date_ntext = soup.find("time").get_text().split(" ")
        if len(date_ntext) == 5:
            date = " ".join(date_ntext[:3])
        else:
            date = " ".join(date_ntext[2:5])
        top_5 = list(map(lambda x: x.get_text().split(".")[1].split(",")[0].strip(), soup.find_all("span", attrs = {"style": "font-size: 18pt;"})[:5]))
        if len(top_5) == 0:
            top_five = soup.find_all(lambda tag: tag.name == 'h3' and "<strong>" in str(tag))
            top_five = BeautifulSoup(str(top_5), "html.parser").get_text().split("\n")
            top_5 = []
            for item in top_five:
                
                name = item.split(",")
                if len(name) > 1:
                     name = name[1].strip()
                top_5.append(name)
        top_6_10 = soup.find_all(lambda tag: tag.name == 'p' and "<strong>6.</strong>" in str(tag))
        if len(top_6_10) == 0:
            top_6_10 = soup.find_all(lambda tag: tag.name == 'p' and "<em>" in str(tag))
        top_6_10 = BeautifulSoup(str(top_6_10), "html.parser").get_text().split("\n")
        top_6_10 = list(map(lambda x: " ".join(x.split(" ")[1:3]).strip(","), top_6_10)) 
        ranking = [i for i in range(1, 11)]
        top_10 = top_5 + top_6_10

        df_ranking = pd.DataFrame({"title": title, "date": date, "ranking": ranking, "player": top_10})
        all_top_ten.append(df_ranking)
    except:
        print(links)
```

```python
part_top_ten = pd.concat(all_top_ten)
```

```python
import pandas as pd
from datetime import datetime

def add_new_column(df_player, df_seasons):
    df_player['date'] = pd.to_datetime(df_player['date'])
    
    # Create an empty column for the season
    df_player['season'] = ''
    
    # Loop through each season in df_seasons
    for i, season_row in df_seasons.iterrows():
        duration_str = season_row['duration'].replace('–', '-')  # Replace en-dashes with hyphens
        try:
            season_start = datetime.strptime(duration_str.split('-')[0].strip(), "%B %d, %Y")
            season_end = datetime.strptime(duration_str.split('-')[1].strip(), "%B %d, %Y")
        except: 
            season_start = datetime.strptime(duration_str.split('-')[0].strip() + ", 1998", "%B %d, %Y")
            season_end = datetime.strptime(duration_str.split('-')[1].strip(), "%B %d, %Y")

        # Update the season column for rows that fall within this season
        mask = (df_player['date'] >= season_start) & (df_player['date'] <= season_end) & (df_player['season'] == '')
        df_player.loc[mask, 'season'] = season_row['season']
    
    return df_player
```

```python
def add_new_column(df_player, df_seasons):
    df_player['date'] = pd.to_datetime(df_player['date'])
    
    # Create an empty column for the season
    df_player['season'] = ''
    
    # Loop through each season in df_seasons
    for i, season_row in df_seasons.iterrows():
        season_start = datetime.strptime(season_row['duration'].split(' – ')[0], "%B %d, %Y")
        season_end = datetime.strptime(season_row['duration'].split(' – ')[1], "%B %d, %Y")
        
        # Get the rows in df_player that fall within this season
        df_rankings_within_season = df_player[(df_player['date'] >= season_start) & (df_player['date'] <= season_end)]
        
        # Update the season column for these rows
        df_rankings_within_season['season'] = season_row['season']
        
        # Update the original df_player with the new season column values
        df_player.update(df_rankings_within_season['season'])
    
    return df_player
```
