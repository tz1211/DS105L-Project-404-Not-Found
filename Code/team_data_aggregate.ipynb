```python
from google.colab import drive
import pandas as pd
```

```python
drive.mount('/content/drive')
df_list = []
eastern_conference_team = ['Cleveland Cavaliers','Toronto Raptors','Detroit Pistons','Orlando Magic','Brooklyn Nets','Boston Celtics','Washington Wizards','Chicago Bulls','New Jersey Nets','Philadelphia 76ers','Atlanta Hawks','Charlotte Hornets','Charlotte Bobcats','Milwaukee Bucks','Indiana Pacers','New York Knicks','Miami Heat']
for year_number in range(2010,2024):
  file_path = f'/content/drive/MyDrive/new_team_conference./{year_number}_team_conference.csv'
  df = pd.read_csv(file_path)
  column_season = 'season'
  column_won_rate = 'win_rate'
  win_rate = df['W']/(df['W']+df['L'])
  df[column_season] = f'{year_number}'
  df[column_won_rate] = win_rate
  df['Conference'] = 'None'
  for n in range(len(df)):
    if df['Unnamed: 0'][n] in eastern_conference_team:
      df.at[n,'Conference'] = 'Eastern'
    else:
      df.at[n,'Conference'] = 'Western'

  df_list.append(df)

merged_df = pd.concat(df_list, axis=0, ignore_index=True)

merged_df = merged_df.rename(columns={'Unnamed: 0':'Team'})

merged_df = merged_df[['season','Team','Conference','W','L','win_rate']]
merged_df = merged_df.sort_values(by=['season','Conference','win_rate'], ascending=False)
merged_df = merged_df.reset_index(drop=True)

merged_df['conference_standing'] = merged_df.groupby(['season','Conference']).cumcount()+1

display(merged_df)
merged_df.to_csv('/content/drive/MyDrive/DS105L group project/team_conference/merged_team_conference.csv', index = False)
```
