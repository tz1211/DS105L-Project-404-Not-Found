```python
from IPython.display import Image
import pydot

# Create the graph object
graph = pydot.Dot(graph_type='digraph')

# Add the nodes to the graph
#contributor = pydot.Node('Contributor -- Terry Zhou,Spark Lu,Annabel Zhong',shape='box',style='filled',fillcolor='lightyellow')
step1 = pydot.Node('1.Data Collection', shape='box', style='filled', fillcolor='yellow')
step2 = pydot.Node('2.Data Cleaning', shape='box', style='filled', fillcolor='yellow')
step3 = pydot.Node('3.Show data', shape='box', style='filled', fillcolor='yellow')
step4 = pydot.Node('4.Machine Learning and Prediction', shape='box', style='filled', fillcolor='yellow')
file1 = pydot.Node('merged_box_data.csv',shape='hexagon',style='filled',fillcolor='lightblue')
file2 = pydot.Node('merged_advanced_data.csv',shape='hexagon',style='filled',fillcolor='lightblue')
cleaned_file1 = pydot.Node('player_data.csv',shape='hexagon',style='filled',fillcolor='lightblue')
cleaned_file2 = pydot.Node('player_data_2023.csv',shape='hexagon',style='filled',fillcolor='lightblue')

#graph.add_node(contributor)
graph.add_node(step1)
graph.add_node(step2)
graph.add_node(step3)
graph.add_node(step4)
graph.add_node(file1)
graph.add_node(file2)
graph.add_node(cleaned_file1)
graph.add_node(cleaned_file2)

# Add the edges to the graph (same as before)
edge1 = pydot.Edge('https://jaebradley.github.io/basketball_reference_web_scraper/api/', 'https://www.basketball-reference.com/', label='API usage tutorial')
edge2 = pydot.Edge(file1, step1,label='Original data for use')
edge3 = pydot.Edge(step1, step2)
edge4 = pydot.Edge('https://www.basketball-reference.com/',file1,label='Scrape data')
edge5 = pydot.Edge('https://www.basketball-reference.com/',file2,label='Scrape data')
edge6 = pydot.Edge(step2, cleaned_file1)
edge7 = pydot.Edge(step3, step4)
edge8 = pydot.Edge('https://www.nba.com/players',step1,label='Scrape Media Opinion')
edge9 = pydot.Edge('https://www.bbc.co.uk/',step1,label='Scrape Media Opinion')
edge10= pydot.Edge(file2,step1,label='Original data for use')
edge11= pydot.Edge(step2, cleaned_file2)
edge12= pydot.Edge(cleaned_file1,step3)
edge13= pydot.Edge(cleaned_file2,step3)

graph.add_edge(edge1)
graph.add_edge(edge2)
graph.add_edge(edge3)
graph.add_edge(edge4)
graph.add_edge(edge5)
graph.add_edge(edge6)
graph.add_edge(edge7)
graph.add_edge(edge8)
graph.add_edge(edge9)
graph.add_edge(edge10)
graph.add_edge(edge11)
graph.add_edge(edge12)
graph.add_edge(edge13)

# Generate and display the graph
Image(graph.create_png())
```
