```python
def get_player_news(one_season):
    base_url = 'https://www.bbc.co.uk/search'
    date_l = []
    url_l = []
    title_l = []
    name_l = []
        
    for player in one_season:
            print(player)
            results = []
            page_number = 1
            title_1 = []

            url_player = player.replace(" ", "-")
            while True:
                if page_number > 29:
                     break
                params = {'q': url_player  + '+mvp', 'd': 'SPORT_PS', 'page': str(page_number)}

                try:
                    response = requests.get(base_url, params=params)
                    soup = BeautifulSoup(response.text, 'html.parser')
                    event_cards = soup.find_all('div', attrs={"data-testid": "default-promo", "class":"ssrcss-53phst-Promo ett16tt0"})

                    if not event_cards:
                        break

                    for card in event_cards:
                        title = card.find('p', attrs={'class': "ssrcss-6arcww-PromoHeadline e1f5wbog5"}).text.strip()
                        title_1.append(title)
                        date = card.find('span', attrs={'class': 'ssrcss-1if1g9v-MetadataText ecn1o5v1'}).text.strip()
                        date_l.append(date)
                        url = card.find('a').get('href')
                        url_l.append(url)
                        results.append({'title': title, 'date': date, 'url': url})

                    page_number += 1
                    print(page_number)

                except Exception as e:
                    break
            name_l += [player] * len(title_1)
            title_l += title_1
    one_season_df = pd.DataFrame({"title": title_l, "date":date_l, 'url': url_l,"player": name_l })
    return one_season_df
        

    
```

```python
bbc_2023 = get_player_news(list(df_2023["name"]))
```

```python
import pandas as pd
from datetime import datetime

def add_new_column(df_news, df_seasons):
    df_news['time'] = pd.to_datetime(df_news['time'])
    
    # Create an empty column for the season
    df_news['season'] = ''
    
    # Loop through each season in df_seasons
    for i, season_row in df_seasons.iterrows():
                    
        duration_str = season_row['duration'].replace('–', '-').strip()  # Replace en-dashes with hyphens

        
            # Convert the duration string to the same format as the time column in df_news
        start_str, end_str = duration_str.split('-')
        season_start, season_end = None, None  # Default values
        
        try:
            if start_str == "February 5":
                start_str = datetime.strptime(start_str + ", 1998", "%B %d, %Y").strftime('%Y-%m-%d')
                end_str = datetime.strptime(end_str, "%B %d, %Y").strftime('%Y-%m-%d')
                season_start = datetime.strptime(start_str, '%Y-%m-%d')
                season_end = datetime.strptime(end_str, '%Y-%m-%d')

            else:
                start_str = datetime.strptime(start_str, '%B %d, %Y').strftime('%Y-%m-%d')
                end_str = datetime.strptime(end_str, '%B %d, %Y').strftime('%Y-%m-%d')
                season_start = datetime.strptime(start_str, '%Y-%m-%d')
                season_end = datetime.strptime(end_str, '%Y-%m-%d')
        except Exception as e:  # Catch all exceptions
            print(e)
            print(start_str)
            print(end_str)



        # Update the season column for rows that fall within this season
        mask = (df_news['time'] >= season_start) & (df_news['time'] <= season_end) & (df_news['season'] == '')
        df_news.loc[mask, 'season'] = season_row['season']
    
    return df_news
```
