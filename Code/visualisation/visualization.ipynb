```python
from IPython.display import Image
import pydotplus

# Create the graph object
graph = pydot.Dot(graph_type='digraph')

# Add the nodes to the graph
step1 = pydot.Node('1.Data Collection', shape='box')
step2 = pydot.Node('2.Show Data', shape='box')
step3 = pydot.Node('3.Data Cleaning', shape='box')
step4 = pydot.Node('4.Machine Learning and Prediction', shape='box')

graph.add_node(step1)
graph.add_node(step2)
graph.add_node(step3)
graph.add_node(step4)

# Add the edges to the graph (same as before)
edge1 = pydot.Edge('https://jaebradley.github.io/basketball_reference_web_scraper/api/', step1)
edge2 = pydot.Edge('https://www.basketball-reference.com/', step1)
edge3 = pydot.Edge(step1, step2)
edge4 = pydot.Edge('merged_box_data.csv', step2)
edge5 = pydot.Edge('merged_advanced_data.csv', step2)
edge6 = pydot.Edge(step2, step3)
edge7 = pydot.Edge(step3, step4)

graph.add_edge(edge1)
graph.add_edge(edge2)
graph.add_edge(edge3)
graph.add_edge(edge4)
graph.add_edge(edge5)
graph.add_edge(edge6)
graph.add_edge(edge7)

# Generate and display the graph
Image(graph.create_png())
```
