```python

def filter_news_by_time(news_df, date_duration):
    # Convert 'date' column to datetime format
    news_df['time'] = pd.to_datetime(news_df['time'], format="%B %d, %Y")
    start_string, end_string = date_duration.split(" – ")
    start_string = start_string.strip(_)
    end_string = end_string.strip()
    start_date = datetime.strptime(start_string, "%B %d, %Y")
    end_date = datetime.strptime(end_string, "%B %d, %Y")
    mask = (news_df['time'] >= start_date) & (news_df['time'] <= end_date)
    news_within_duration = news_df.loc[mask]


    return news_within_duration
```

```python
import pandas as pd
from datetime import datetime

def add_new_column(df_news, df_seasons):
    df_news['time'] = pd.to_datetime(df_news['time'], format='%B %d, %Y')
    
    df_news['season'] = ''
    
    # Loop through each season in season dataframe
    for i, season_row in df_seasons.iterrows():
        duration_str = season_row['duration'].replace('–', '-')  # replace dashes with hyphens
        try:
            season_start = datetime.strptime(duration_str.split('-')[0].strip(), "%B %d, %Y")
            season_end = datetime.strptime(duration_str.split('-')[1].strip(), "%B %d, %Y")
        except: 
            season_start = datetime.strptime(duration_str.split('-')[0].strip() + ", 1998", "%B %d, %Y")
            season_end = datetime.strptime(duration_str.split('-')[1].strip(), "%B %d, %Y")
        
        mask = (df_news['time'] >= season_start) & (df_news['time'] <= season_end) & (df_news['season'] == '')
        df_news.loc[mask, 'season'] = season_row['season']
    
    return df_news
```

```python
def find_author(df_news):
    list_url = list(df_news["url"])
    author_list = []
    for url in list_url:
        try:
            r = requests.get(url) 
            soup = BeautifulSoup(r.content, "html.parser") 
            author = soup.find("span", attrs ={"class": "entry-header__author"}).contents[0]
            author_list.append(author)
        except:
            author = "NA"
            author_list.append(author)
    df_news["author"] = author_list
    print(url)
    return df_news 


    
```
