```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
from nba_api.stats.endpoints import playercareerstats
import json
import pandas as pd
from google.colab import drive
import time
```

```python
for year in range(2010,2024):
  
  PLAYER_JSON = client.players_season_totals(season_end_year=2010, output_type=OutputType.JSON)
  total_t = json.loads(PLAYER_JSON)

  PLAYER_stats = pd.json_normalize(total_t)

  # get the players who play more than 30 minutes

  Player_list_1 = PLAYER_stats[PLAYER_stats['minutes_played']/PLAYER_stats['games_played'] >= 30]

  new_columns = ['name'] + ['minutes_played'] + ['games_played'] + ['points'] + ['team'] + [c for c in Player_list_1.columns if c not in ('name','minutes_played','games_played','points','team')]
  Player_list_1 = Player_list_1[new_columns]

  #print(Player_list_1.columns)
  Player_list_1 = Player_list_1.sort_values(by='points', ascending=False)
  display(Player_list_1)

  time.sleep(20)

  drive.mount('/drive')
  Player_list_1.to_csv('/drive/My Drive/data/2010_player_data.csv')
```

```python
for year in range(2010,2024):
  PLAYER_advanced_JSON = client.players_advanced_season_totals(season_end_year=year, output_type=OutputType.JSON)
  advanced_total_t = json.loads(PLAYER_advanced_JSON)

  PLAYER_advanced_stats = pd.json_normalize(advanced_total_t)
  advanced_new_column = ['name'] + [c for c in PLAYER_advanced_stats.columns if c != 'name']
  PLAYER_advanced_stats = PLAYER_advanced_stats[advanced_new_column]
  PLAYER_advanced_stats = PLAYER_advanced_stats.sort_values(by='player_efficiency_rating', ascending=False)
  display(PLAYER_advanced_stats)
  time.sleep(20)

  drive.mount('/drive')
  PLAYER_advanced_stats.to_csv(f'/drive/My Drive/advanced_data/{year}_player_data.csv')
```

```python
for year in range(2022,2023):
  a = client.season_schedule(season_end_year=year, output_type=OutputType.JSON)
  schedule_json = json.loads(a)

  team_conference = {}

  for each_row in schedule_json:
    if each_row['away_team_score'] > each_row['home_team_score']:

      if each_row['away_team'] not in team_conference:
        team_conference[each_row['away_team']] = {'W':1,'L':0}
        if each_row['home_team'] not in team_conference:
          team_conference[each_row['home_team']] = {'W':0,'L':1}
        else:
          team_conference[each_row['home_team']]['L'] += 1

      elif each_row['away_team'] in team_conference:
        team_conference[each_row['away_team']]['W'] += 1
        if each_row['home_team'] not in team_conference:
          team_conference[each_row['home_team']] = {'W':0,'L':1}
        else:
          team_conference[each_row['home_team']]['L'] += 1
        

    elif each_row['away_team_score'] < each_row['home_team_score']:

      if each_row['home_team'] not in team_conference:
        team_conference[each_row['home_team']] = {'W':1,'L':0}
        if each_row['away_team'] not in team_conference:
          team_conference[each_row['away_team']] = {'W':0,'L':1}
        else:
          team_conference[each_row['away_team']]['L'] += 1

      elif each_row['home_team'] in team_conference:
        team_conference[each_row['home_team']]['W'] += 1
        if each_row['away_team'] not in team_conference:
          team_conference[each_row['away_team']] = {'W':0,'L':1}
        else:
          team_conference[each_row['away_team']]['L'] += 1

    team_conference_dataframe = pd.DataFrame(team_conference).T

    team_conference_dataframe = team_conference_dataframe.sort_values(by='W',ascending=False)
    display(team_conference_dataframe)

    time.sleep(60)

    drive.mount('/drive')
    team_conference_dataframe.to_csv(f'/drive/MyDrive/DS105L group project/team_conference/{year}_team_conference.csv')
```
