```python
import os
import pandas as pd
from google.colab import drive

# set directory path
directory = "/content/drive/MyDrive/data"

# create an empty list to store DataFrames
df_list = []

# loop through each file in the directory
for file in os.listdir(directory):
    # check if file is a CSV file
    if file.endswith(".csv"):
        # read CSV file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file))
        # add file name as a column to the DataFrame
        df.insert(0, "year", file)
        # append DataFrame to list
        df_list.append(df)

# concatenate all DataFrames in the list into a single DataFrame
merged_df = pd.concat(df_list, ignore_index=True)

# merge cells in the "year" column by the same index
merged_df["year"] = merged_df.groupby(merged_df.index)["year"].agg('first')

# print the resulting DataFrame
display(merged_df)


drive.mount('/drive')
merged_df.to_csv('/drive/My Drive/data/merged_player_data_2010_2023.csv')


```
