```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import RandomForestClassifier 
```

# Import Data 

```python
df_final = pd.read_csv("../Data/player_data.csv")
df_final.drop("Unnamed: 0", axis = 1, inplace = True) 
df_final
```

```python
df_2023 = pd.read_csv("../Data/player_data_2023.csv")
df_2023.drop("Unnamed: 0", axis = 1, inplace = True) 
df_2023
```

# Random Forest Algorithm 

## Back testing using historic data 

```python
def predict_mvp(df_id, train_x, train_y, test_x, display = 5, to_print=False): 
    random_forest = RandomForestClassifier() 
    random_forest.fit(train_x, train_y) 
    probs = random_forest.predict_proba(test_x) 
    mvp_prob = probs[:, 1] 
    mvp_prob = np.vectorize(lambda x: x/sum(mvp_prob))(mvp_prob)

    df_id["probability"] = mvp_prob 
    df_results = df_id.sort_values("probability", ascending=False)
    df_results = df_results[df_results["probability"] > 0]

    if to_print: 
        print(df_results.iloc[:display, :])

    return random_forest
```

```python
for season in range(2010, 2023): 
    df_test = df_final[df_final["season"] == season] 
    df_train = df_final[df_final["season"] != season] 
    df_id = df_test.iloc[:, :5]
    df_id.drop("slug", axis = 1, inplace=True)

    train_x = df_train.iloc[:, 5:-1]
    train_y = df_train.iloc[:, -1] 
    test_x = df_test.iloc[:, 5:-1]

    predict_mvp(df_id, train_x, train_y, test_x, to_print=True)
```

## Predicting 2023 MVP 

```python
df_id = df_2023.iloc[:, :5]
df_id.drop("slug", axis = 1, inplace=True)

train_x = df_final.iloc[:, 5:-1]
train_y = df_final.iloc[:, -1] 
test_x = df_2023.iloc[:, 5:]
```

```python
rf_model = predict_mvp(df_id, train_x, train_y, test_x, to_print=True)
```

## Feature Importance 

```python
feature_importances = pd.DataFrame({"var": train_x.columns, "importance": rf_model.feature_importances_}).sort_values('importance', ascending=False)
feature_importances
```
