```python
import pandas as pd 
```

```python
df_all_bbc = []

for i in range(10, 24): 
    df_partial = pd.read_csv(f"../../Data/news_final_data/bbc_20{i}_unique.csv") 
    df_all_bbc.append(df_partial) 

len(df_all_bbc)
```

```python
df_all_nba = [] 

for i in range(16, 21): 
    df_partial = pd.read_csv(f"../../Data/news_final_data/nba_20{i}_unique.csv") 
    df_all_nba.append(df_partial) 

len(df_all_nba)
```

```python
df_bbc = pd.concat(df_all_bbc)
df_bbc.drop("url", axis = 1, inplace = True) 
df_bbc.rename(columns={
    "time": "season", 
    "most_frequent_name": "name"
}, inplace = True)
df_bbc
```

```python
df_nba = pd.concat(df_all_nba)
df_nba.drop(["Unnamed: 0", "url"], axis = 1, inplace = True) 
df_nba.rename(columns={
    "time": "season", 
    "most_frequent_name": "name" 
}, inplace = True)
df_nba[df_nba["frequency"] >= 20]
```

```python
df_bbc_2023 = df_bbc[df_bbc["season"] == 2023]
df_bbc = df_bbc[df_bbc["season"] != 2023]
```
