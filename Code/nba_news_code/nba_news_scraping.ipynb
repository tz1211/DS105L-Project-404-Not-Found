```python
import requests 
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://global.nba.com/news/") 
r.status_code
```

```python
articles, time = [], []
for page in range(1201, 1641): #Break the pages down into 3 scrapes so it doesn't time out and combine the df afterwards
    r = requests.get(f"http://global.nba.com/news/page/{page}") 
    soup = BeautifulSoup(r.content, "html.parser") 
    articles += soup.find_all("h1", attrs = {"class": "entry-header__title"}) 
    time += soup.find_all("time") 
    if page % 10 == 0: #Monitoring progress
        print(f"page {page}")
```

```python
df = pd.DataFrame({"title": articles, "time": time}) 
df["url"] = df["title"].apply(lambda x: x.find("a").get("href"))
df["title"] = df["title"].apply(lambda x: x.get_text().strip())
df["time"] = df["time"].apply(lambda x: x.get_text())
df
```

```python
df_1 = pd.read_csv("nba_news_archive_600.csv") 
df_2 = pd.read_csv("nba_news_archive_600_1200.csv") 
df_3 = pd.read_csv("nba_news_archive_1200_1640.csv") 
```

```python
df = pd.concat([df_1, df_2, df_3])
df.drop(columns = "Unnamed: 0", inplace = True)
df = df[df["title"].notna()]
df.reset_index(drop = True, inplace = True)
df
```

```python
df.to_csv("nba_news_archive.csv")
```
