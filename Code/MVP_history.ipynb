```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import itertools
from google.colab import drive
```

```python
drive.mount('/content/drive')
url = 'http://www.espn.com/nba/history/awards/_/id/33'
response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')
table = soup.find('table',{'class':'tablehead'},{'cellspacing':'0'})

table_data = pd.read_html(str(table))

df = table_data[0]
display(df)

year_list = list(df.iloc[2:69,0])
new_year_list = []
for years in year_list:
  new_year_list.append(int(years))
MVP_list = list(df.iloc[2:69,1])
result_value = [0]*4489
combination = itertools.product(MVP_list, new_year_list)
MVP_dataframe = pd.DataFrame(combination, columns=['Name', 'Year'])
MVP_dataframe['value'] = result_value

MVP_dataframe.insert(loc=3, column='won_last_yr', value=0)
MVP_dataframe.insert(loc=4, column='won_2_yrs_before', value=0)

for number_index in range(0,67):
  mvp_name = MVP_list[number_index]
  target_year = new_year_list[number_index]
  MVP_dataframe.loc[(MVP_dataframe['Name'] == mvp_name) & (MVP_dataframe['Year'] == target_year), 'value'] += 1
  target_first_cell = mvp_name
  target_second_cell = target_year + 1
  target_third_cell = target_year + 2
  MVP_dataframe.loc[(MVP_dataframe['Name'] == target_first_cell) & (MVP_dataframe['Year'] == target_second_cell), 'won_last_yr'] = 1
  MVP_dataframe.loc[(MVP_dataframe['Name'] == target_first_cell) & (MVP_dataframe['Year'] == target_third_cell), 'won_2_yrs_before'] = 1

MVP_dataframe.drop_duplicates(subset=['Name', 'Year'], keep='first', inplace=True)

display(MVP_dataframe)
MVP_dataframe.to_csv('/content/drive/MyDrive/DS105L group project/MVP_history.csv',index=False)
```
