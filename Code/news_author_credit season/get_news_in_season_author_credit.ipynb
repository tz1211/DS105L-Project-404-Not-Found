```python
import pandas as pd
from datetime import datetime

def add_new_column(df_news, df_seasons):
    df_news['time'] = pd.to_datetime(df_news['time'])
    
    # Create an empty column for the season
    df_news['season'] = ''
    
    # Loop through each season in df_seasons
    for i, season_row in df_seasons.iterrows():
                    
        duration_str = season_row['duration'].replace('–', '-').strip()  # Replace en-dashes with hyphens

        try:
            # Convert the duration string to the same format as the time column in df_news
            start_str, end_str = duration_str.split('-')
            start_str = datetime.strptime(start_str, '%B %d, %Y').strftime('%Y-%m-%d')
            end_str = datetime.strptime(end_str, '%B %d, %Y').strftime('%Y-%m-%d')
            season_start = datetime.strptime(start_str, '%Y-%m-%d')
            season_end = datetime.strptime(end_str, '%Y-%m-%d')
        except ValueError: 

            start_str = datetime.strptime(str(duration_str.split('-')[0]) + ", 1998", "%B %d, %Y").strftime('%Y-%m-%d')
            end_str = datetime.strptime(duration_str.split('-')[1], "%B %d, %Y").strftime('%Y-%m-%d')
            season_start = datetime.strptime(start_str, '%Y-%m-%d')
            season_end = datetime.strptime(end_str, '%Y-%m-%d')

        # Update the season column for rows that fall within this season
        mask = (df_news['time'] >= season_start) & (df_news['time'] <= season_end) & (df_news['season'] == '')
        df_news.loc[mask, 'season'] = season_row['season']
    
    return df_news
```

```python
# function which finds the news contain player
pd_all_news_1 = df_nba_news_archive.copy()
pd_all_news_2 = pd_all_news_1.copy()

def get_player_news(df_news, player_list):
    player_news_list = []
    for player in player_list:
        try:
            pd_player = df_news[df_news["title"].str.contains(str(player), na = False)]
            pd_key = pd_player[pd_player["title"].str.contains("MVP", na = False)]
            rows = pd_key.shape[0]
            player = [player] * rows
            pd_key["player"] = player
            player_news_list.append(pd_key)
        except:
            print(player)

    final_df = pd.concat(player_news_list, ignore_index=True)
    #final_df.drop(columns = "Unnamed: 0", inplace = True)
    #final_df = final_df[final_df["title"].notna()]
    #final_df.reset_index(drop = True, inplace = True)
    return final_df
```

# find author

```python
def find_author(df_news):
    list_url = list(df_news["url"])
    author_list = []
    for url in list_url:
        try:
            r = requests.get(url) 
            soup = BeautifulSoup(r.content, "html.parser") 
            author = soup.find("span", attrs ={"class": "entry-header__author"}).contents[0]
            author_list.append(author)
        except:
            author = "NA"
            author_list.append(author)
    df_news["author"] = author_list
    return df_news 


    
```
