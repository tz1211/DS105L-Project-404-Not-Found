```python
def get_bbc_textblob_sentiment(df):
    sentiment_scores = []

    for index, row in df.iterrows():
        url = row['url']
        title = row['title']

        r = requests.get(url)
        soup = BeautifulSoup(r.content, "html.parser") 
        content = soup.find("div", attrs={"class": "gel-layout__item gel-2/3@l"})

        if content is None:
            content = soup.find("td", attrs={"class": "storybody"})
        elif content is None:
            clean_text = title
        else:
            main_body = content.find_all('p')
            text = ""
            for i in main_body:
                text += i.text
            clean_text = text.replace('\n', ' ').replace('\xa0', '')

        blob = TextBlob(clean_text)
        sentiment = blob.sentiment
        polarity = sentiment.polarity  # Assess sentiment polarity (-1 to 1)
        sentiment_scores.append(polarity)

    df['textblob_sentiment'] = sentiment_scores
    return df
```

```python
def assess_sentiment_nltk(text):
    sentiment_analyzer = SentimentIntensityAnalyzer()
    sentiment_scores = sentiment_analyzer.polarity_scores(text)
    score = sentiment_scores['compound']
    return score

def get_bbc_nltk_sentiment(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser") 
    content = soup.find("div", attrs={"class": "gel-layout__item gel-2/3@l"})
    
    if content is None:
        content = soup.find("td", attrs={"class": "storybody"})
        if content is None:
            return None

    main_body = content.find_all('p')
    text = ""
    for i in main_body:
        text += i.text
    clean_text = text.replace('\n', ' ').replace('\xa0', '')

    score = assess_sentiment_nltk(clean_text)
    return score

```
